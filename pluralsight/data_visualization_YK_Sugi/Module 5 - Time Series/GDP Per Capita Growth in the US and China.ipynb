```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
```

```python
data.head()
```

```python
us = data[data.country == 'United States']
ch = data[data.country == 'China']
```

```python
plt.plot(us.year, us.gdpPerCapita)
plt.plot(ch.year, ch.gdpPerCapita)
plt.xlabel('year')
plt.ylabel('GDP Per Capita')
plt.title('GDP Per Capita per Year')
plt.legend(['US', 'CH'])
plt.show()
```

```python
us.gdpPerCapita.iloc[0]
```

```python
us_growth = us.gdpPerCapita / us.gdpPerCapita.iloc[0] * 100
ch_growth = ch.gdpPerCapita / ch.gdpPerCapita.iloc[0] * 100
```

```python
plt.plot(us.year, us_growth)
plt.plot(ch.year, ch_growth)
plt.xlabel('year')
plt.ylabel('GDP Per Capita Growth')
plt.title('GDP Per Capita Growth per Year (first year = 100)')
plt.legend(['US', 'CH'])
plt.show()
```
