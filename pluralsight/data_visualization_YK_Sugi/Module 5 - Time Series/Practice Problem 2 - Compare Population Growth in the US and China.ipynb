```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('countries.csv')
data.head()
```

```python
us = data[data.country == 'United States']
ch = data[data.country == 'China']
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(ch.year, ch.population / 10**6)
plt.legend(['US', 'CH'])
plt.xlabel('year')
plt.ylabel('population')
plt.title('Absolute Population Growth per Year (in millions)')
plt.show()
```

```python
us_growth = us.population / us.population.iloc[0] * 100
ch_growth = ch.population / ch.population.iloc[0] * 100
```

```python
plt.plot(us.year, us_growth)
plt.plot(ch.year, ch_growth)
plt.legend(['US', 'CH'])
plt.xlabel('year')
plt.ylabel('population')
plt.title('Relative Population Growth per Year (first year = 100)')
plt.show()
```
