```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('../countries.csv')
data.head()
```

```python
data_2007 = data[data.year == 2007]
data_2007.head()
```

```python
plt.scatter(data_2007.gdpPerCapita, data_2007.lifeExpectancy, 10)
plt.title('GDP Per Capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
data_2007.gdpPerCapita.corr(data_2007.lifeExpectancy)
```

```python
np.log10([10, 100, 1000])
```

```python
plt.scatter(np.log10(data_2007.gdpPerCapita), data_2007.lifeExpectancy)
plt.title('GDP Per Capita and Life Expectancy in 2007')
plt.xlabel('GDP Per Capita ($) (log scale)')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
np.log10(data_2007.gdpPerCapita).corr(data_2007.lifeExpectancy)
```

```python
years_sorted = sorted(set(data.year))

for year in years_sorted:
    data_year = data[data.year == year]
    plt.scatter(data_year.gdpPerCapita, data_year.lifeExpectancy, 5)
    plt.title(year)
    plt.xlabel('GDP Per Capita')
    plt.ylabel('Life Expectancy')
    plt.xlim(0, 60000)
    plt.ylim(25, 85)
    plt.show()
    # plt.savefig(str(year), dpi=200) # dpi (dots per inch) = resolution of the graph
    # plt.clf() # Clear the current plot to be used in the next loop
```

```python
years_sorted = sorted(set(data.year))

for year in years_sorted:
    data_year = data[data.year == year]
    plt.scatter(np.log10(data_year.gdpPerCapita), data_year.lifeExpectancy, 5)
    plt.title(year)
    plt.xlabel('GDP Per Capita (log scale)')
    plt.ylabel('Life Expectancy')
    plt.xlim(2, 5)
    plt.ylim(25, 85)
    plt.show()
    # plt.savefig('log_' + str(year), dpi=200) # dpi (dots per inch) = resolution of the graph
    # plt.clf() # Clear the current plot to be used in the next loop
```
